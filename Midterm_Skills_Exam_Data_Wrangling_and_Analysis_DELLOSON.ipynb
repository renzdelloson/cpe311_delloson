```python
c_data['workclass'].replace('?', 'Private', inplace=True)
c_data['workclass'].fillna('Private', inplace=True)
c_data['native-country'].replace('?', 'United-States', inplace=True)
c_data['native-country'].fillna('United-States', inplace=True)
c_data['occupation'].replace('?', np.NaN, inplace=True)
```

```python
c_data.replace('<=50K.', '<=50K', inplace=True)
c_data.replace('>50K.', '>50K', inplace=True)
```

```python
c_data = c_data.assign(
    occupation = lambda x: x['occupation'].fillna(method='ffill')
)
```

```python
census_cat = c_data.copy()
```

```python
# To remove 'education' because education already has an education-num. dropping 'education' would be helpful.
c_data.drop(columns=['education'], inplace=True)
c_data
```

```python
def preprocessing(data, catlist):
    if data[catlist].dtype == 'object':
        categorical_values = data[catlist].unique()
        range_values = range(1, len(categorical_values) + 1)
        mapping = {val: idx for val, idx in zip(categorical_values, range_values)}
        print(f"{catlist}:", mapping)
        data[catlist] = data[catlist].map(mapping)
    return data

for column in c_data.select_dtypes(include=['object']).columns:
    c_data = preprocessing(c_data, column)
```

```python
# creating a heatmap to examine the correlation among different categories
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 12))
sns.heatmap(
    c_data.corr(),
    annot=True,
    cmap='coolwarm',
    linewidths=.5,
    linecolor='black',
    cbar_kws={'orientation': 'horizontal'}
)
plt.title('Correlation Heatmap', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sampl_cen = census_cat.sample(300, random_state=0)
sampl_cen.groupby('workclass')['age'].describe()
```

```python
# Making Data Plots and Conducting Data Analysis
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(4, figsize = [10,20])

# marital-status column
c_data.groupby('relationship').size().plot(kind='barh', ax = ax[0], color = ('yellow','green'))
ax[0].set_title('Relationship of the Individuals')
ax[0].set_xlabel('Number of Individuals')


# relationship column
c_data.groupby('marital-status').size().plot(kind='barh', ax = ax[1], color = ('pink','white'))
ax[1].set_title('Martial Status of the Individuals')
ax[1].set_xlabel('Number of Individuals')


# sex column
c_data.groupby('sex').size().plot(kind='barh', ax = ax[2], color = ('blue','brown'))
ax[2].set_title('Gender of the Individuals')
ax[2].set_xlabel('Number of Individuals')


# race column
c_data.groupby('race').size().plot(kind='barh', ax = ax[3], color = ('skyblue','black'))
ax[3].set_title('Race Individuals')
ax[3].set_xlabel('Number of Individuals')


fig.tight_layout()
```

```python
# plotting of the age and workclass with the use of boxplot
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(15, 5))
sns.boxplot(data=sampl_cen, x='sex', y='occupation', hue='occupation', ax=ax, palette='Set2')
ax.set_title('Male and Female in Occupation', fontsize=20, fontweight='semibold', color='darkblue')

ax.set_xticklabels(ax.get_xticklabels(), fontsize=11, rotation=45, ha='right')

ax.set_yticklabels(ax.get_yticks(), fontsize=11)

ax.legend(title='Occupation', title_fontsize='13', fontsize='11', loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# plotting of the Income of each individuals per workclass with the use of stacked bar graphs
crosstab = pd.crosstab(census_cat['occupation'], census_cat['hours-per-week'])
fig, ax = plt.subplots(figsize=(25,70))
crosstab.plot(kind='bar', stacked=True, ax=ax, color=['red', 'yellow'])
ax.set_title('Hours per week of each person in different Occupation', fontsize=18, fontweight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=90)
ax.set_yticklabels(ax.get_yticks(), fontsize=12)
ax.legend(fontsize=16)
```

```python
# plotting of the education attainment of each individuals per workclass with the use of stacked bar graphs
crosstab = pd.crosstab(census_cat['occupation'], census_cat['capital-gain'])
fig, ax = plt.subplots(figsize=(25,60))
crosstab.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Capital Gain of individuals in different Occupation', fontsize=18, fontweight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=16, rotation=90)
ax.set_yticklabels(ax.get_yticks(), fontsize=16)
ax.legend(fontsize=16)
```

```python
# plotting of the Income of each individuals per workclass with the use of stacked bar graphs
crosstab = pd.crosstab(census_cat['occupation'], census_cat['race'])
fig, ax = plt.subplots(figsize=(20,60))
crosstab.plot(kind='bar', stacked=True, ax=ax, color=['pink', 'blue', 'green', 'violet', 'skyblue'])
ax.set_title('Race of the Individuals', fontsize=22, fontweight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=90)
ax.set_yticklabels(ax.get_yticks(), fontsize=12)
ax.legend(fontsize=22)
```
