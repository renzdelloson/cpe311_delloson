```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pop = pd.read_csv('/content/Listing_of_Active_Businesses.csv')
pop.head()
```

```python
pop.info()
```

```python
# keeps only these data
newPop = pd.concat([
    pop.iloc[:, 0], # BUSINESS NAME
    pop.iloc[:, 1], # DBA NAME
    pop.iloc[:, 2], # STREET ADDRESS
    pop.iloc[:, 7], # CITY
], axis = 1)
newPop
```

```python
newPop.duplicated().sum()
```

```python
# Create a list of Business names
Business_List =  ['KOMODO LLC	', 'RICHARD JOHN SHERMAN', 'SOCAL DEMOLITION COMPANY', 'LDE CONSULTING',
              'GARLAND MFG CO/C	', 'WALTER ARIAS', 'NATALIE JOANNE RICHTER']
```

```python
Business = newPop[newPop['BUSINESS NAME'].isin(Business_List)]
Business.set_index('BUSINESS NAME', inplace = True)
Business
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(data=pop, x='CITY', order=pop['CITY'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of Businesses by City')
plt.xlabel('City')
plt.ylabel('Number of Businesses')
plt.show()
```
