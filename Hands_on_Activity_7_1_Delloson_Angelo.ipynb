This is the source for the hospital information that i get:

 https://sulit.ph/list-of-hospitals-in-metro-manila-with-contact-details-website-and-social-media-accounts/

**So After searching for some website for the hospital list i did find a website for this type of exercise.**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_hospitals_in_the_Philippines"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

tables = soup.find_all("table")

df_list = []

for table in tables:
    headers = [header.get_text(strip=True) for header in table.find_all("th")]


    if "Name of Hospital" in headers and "Location" in headers:
        name_index = headers.index("Name of Hospital")
        location_index = headers.index("Location")

        data_rows = []

        for row in table.find_all("tr")[1:]:
            columns = row.find_all("td")
            if len(columns) >= len(headers):
                name = columns[name_index].get_text(strip=True)
                location = columns[location_index].get_text(strip=True)
                data_rows.append([name, location])

        df = pd.DataFrame(data_rows, columns=["Name of Hospital", "Location"])
        df_list.append(df)


hospitals_df = pd.concat(df_list, ignore_index=True)

hospitals_df.to_csv('/content/hospitals.csv', index=False)

hospitals_df
```
